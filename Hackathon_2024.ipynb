```python
import pandas as pd

# Load the ratings and movies data
ratings = pd.read_csv('/content/drive/MyDrive/Hackathon_2024/movie_data/ratings.csv')
movies = pd.read_csv('/content/drive/MyDrive/Hackathon_2024/movie_data/movies.csv')

# Step 1: Group user ratings based on movieId and apply count and mean
ratings_grouped = ratings.groupby('movieId').agg({'rating': ['count', 'mean']}).reset_index()
ratings_grouped.columns = ['movieId', 'rating_count', 'rating_mean']

# Step 2: Inner join on dataframe created from movies.csv and the grouped df from step 1
merged_df = pd.merge(movies, ratings_grouped, on='movieId', how='inner')

# Step 3: Filter only those movies which have more than 50 user ratings
filtered_movies = merged_df[merged_df['rating_count'] > 50]

# Display the resulting DataFrame
print("Movies with more than 50 user ratings:")
print(filtered_movies)
```

```python
# Find the movie with the highest rating count
max_rating_movie = filtered_movies[filtered_movies['rating_count'] == filtered_movies['rating_count'].max()]

print(max_rating_movie[['movieId']])
```

```python

# Filtering movies with the 'Sci-Fi' genre
scifi_movies = filtered_movies[filtered_movies['genres'].str.contains('Sci-Fi')]

# Finding the Sci-Fi movie with the highest IMDb rating
max_imdb_scifi_movie = scifi_movies[scifi_movies['rating_count'] == scifi_movies['rating_count'].max()]

# Printing the movieId of the Sci-Fi movie with the highest IMDb rating
print(max_imdb_scifi_movie[['movieId']])
```

```python
import requests
import numpy as np
from bs4 import BeautifulSoup
links=pd.read_csv("/content/drive/MyDrive/Hackathon_2024/movie_data/links.csv")
def scrapper(imdbId):
    id = str(int(imdbId))
    n_zeroes = 7 - len(id)
    new_id = "0"*n_zeroes + id
    URL = f"https://www.imdb.com/title/tt%7Bnew_id%7D/"
    request_header = {'Content-Type': 'text/html; charset=UTF-8',
                      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0',
                      'Accept-Encoding': 'gzip, deflate, br'}
    response = requests.get(URL, headers=request_header)
    soup = BeautifulSoup(response.text)
    imdb_rating = soup.find('span', attrs={'class' : 'sc-bde20123-1 cMEQkK'})
    return imdb_rating.text if imdb_rating else np.nan
# final = pd.merge(filtered_movies,links,on='movieId')['imdbId'].apply(scrapper)
```

```python
imdb_rating = pd.merge(filtered_movies,links,on='movieId')
imdb_rating['im_ratings'] = imdb_rating['imdbId'].apply(scrapper)
imdb_rating[imdb_rating['im_ratings'] == imdb_rating['im_ratings'].max()]
```
